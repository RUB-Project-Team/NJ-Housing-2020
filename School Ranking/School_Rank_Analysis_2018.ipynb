```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
import os
```

```python
#File Path
file_path=os.path.join("..","Final Output Data","school_data_2018.csv")
```

```python
# Read School Data File and store into Pandas Data Frames
school_data_df = pd.read_csv("../Final Output Data/schoolRank_data.csv")
school_data_df
```

```python
#checking the data type
school_data_df.dtypes
```

```python
#checking the number of records
school_data_df.count()
```

```python
#Extracting data for school year 2018
school_data_2018=school_data_df.drop(columns=['2019 Rank', '2019 Rank Stars', '2019 Rank Level', '2019 Avg Score',
                                              'School Name', 'Address','Low Grade','High Grade', '2019 Student Count',
                                             ])
school_data_2018
```

```python
#Checking for any null values
school_data_2018.count()
```

```python
#exporting 2018 data to csv
school_data_2018.to_csv(file_path)
```

```python
#top 10 performing schools of 2018
school_data_2018_sorted=school_data_2018.sort_values('2018 Rank')
school_data_2018_sorted.head(10)
```

```python
x=school_data_2018['School Level']
y=school_data_2018['2018 Rank']
plt.scatter(x, y, marker ='o', facecolor='red',edgecolor='blue')
plt.ylim(0, 1400)
plt.show()
```

```python
#Bottom 10 schools of 2018
school_data_2018_bottom=school_data_2018.sort_values('2018 Rank', ascending=False)
school_data_2018_bottom.head(10)
```

```python
avg_rank_zip=school_data_2018.groupby(['Zip Code']).mean()[['2018 Rank','2018 Avg Score']]
avg_rank_zip
```

```python
avg_rank_zip.to_csv("../Final Output Data/avg_rank_zip_2018.csv")
```

```python
avg_rank_city=school_data_2018.groupby(['City']).mean()[['2018 Rank', '2018 Avg Score']]
avg_rank_city
```

```python
avg_rank_city.to_csv("../Final Output Data/avg_rank_city_2018.csv")
```

```python
rank_sizes = school_data_2018.groupby(['2018 Rank Stars']).count()
rank_sizes
```

```python
sizes=rank_sizes['2018 Rank'] 
colors = ['Orange','Red','yellow',"gold","Green","Silver"]
labels=['Zero Star','One Star','Two Star', 'Three Star','Four Star','Five Star']
explode = (0.03,0.03,0.03,0.03,0.03,0.1)
plt.figure(figsize=(5,5), dpi=80)
plt.pie(sizes, labels=labels, colors =colors, autopct='%1.1f%%', explode=explode, startangle =140 )
plt.title("School Rank Distribution in NJ for the Year 2018")
plt.show()
```

```python
# School distribution in NJ by school level-elementary, middle, high, or other
rankLevel_count=school_data_2018.groupby(['2018 Rank Level']).count()
rankLevel_count
```

```python
sizes=rankLevel_count['School Level']
labels = ["Elementary", "Middle", "High", "Other"]
colors=['Green','Red','Yellow','Orange']
explode=(0.03, 0.03, 0.05, 0.2)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)
plt.title("Percentage of Elementary, Middle, High, and Other Schools in NJ")
plt.show
```
